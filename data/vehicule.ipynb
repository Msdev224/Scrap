```python
import pandas as pd
```

```python
# Chemin du fichier
file_path = "brute/expat_voiture.csv"

# Lecture du fichier
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

```python
# Suppression des colonnes inutiles
df_cleaned = df.drop(columns=["web-scraper-order", "web-scraper-start-url", "article-href", 'article'])
```

```python
df_cleaned.head()
```

```python
# Suppression des lignes avec des valeurs manquantes critiques (marque, année, prix)
df_cleaned = df_cleaned.dropna(subset=["marque", "annee", "prix"])
```

```python
# Remplacement des valeurs manquantes
df_cleaned["etat"].fillna("Inconnu", inplace=True)
df_cleaned["boite_vitesse"].fillna("Non spécifié", inplace=True)
df_cleaned["adresse"].fillna("Non précisé", inplace=True)
```

```python
# Conversion de 'annee' en entier
df_cleaned["annee"] = df_cleaned["annee"].astype(int)

# Nettoyage et conversion de la colonne 'prix' en entier
df_cleaned["prix"] = df_cleaned["prix"].str.replace(r"\D", "", regex=True).astype(float)
```

```python
# Réinitialisation de l'index
df_cleaned.reset_index(drop=True, inplace=True)
# Sauvegarde du fichier nettoyé
df_cleaned.to_csv("cleaned/expat_voiture_cleaned.csv", index=False)

# Affichage des premières lignes du dataframe nettoyé
df_cleaned.head()
```

```python
# Fonction de nettoyage des données
def nettoyer_donnees(file_path, output_file):
    df = pd.read_csv(file_path)
    df_cleaned = df.drop(columns=["web-scraper-order", "web-scraper-start-url", "article-href", "scooter"], errors='ignore')
    df_cleaned = df_cleaned.dropna(subset=["marque", "annee", "prix"], how='any')
    df_cleaned.loc[:, "etat"] = df_cleaned["etat"].fillna("Inconnu")
    df_cleaned.loc[:, "adresse"] = df_cleaned["adresse"].fillna("Non précisé")
    df_cleaned.loc[:, "annee"] = df_cleaned["annee"].astype(int)
    df_cleaned.loc[:, "prix"] = df_cleaned["prix"].astype(str).str.replace(r"\D", "", regex=True).astype(int)
    df_cleaned.reset_index(drop=True, inplace=True)
    df_cleaned.to_csv(output_file, index=False)
    print(f"Fichier nettoyé enregistré : {output_file}")
```

```python
#  Nettoyage du fichier scooter
nettoyer_donnees(file_scooter, "cleaned/expat_scooter_cleaned.csv")
```

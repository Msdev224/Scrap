```python
import pandas as pd 
import re
```

```python
file_path =  'brute/expat_equipements.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
columns = ['details', 'etat', 'prix','adresse','image_link-src']
```

```python
df = df[columns]
```

```python
df
```

```python
df.info()
```

```python
df = df.rename(columns={'image_link-src': 'image_link'})
```

```python
def extract_price(price):
    if isinstance(price, str):
        return ''.join(re.findall(r'\d+', price))
    return str(price)
```

```python
df['prix'] = df['prix'].apply(extract_price)
```

```python
df['prix'] = df['prix'].astype(float)
```

```python
df['prix'] = df['prix'].fillna(0)
df['details'] = df['details'].fillna("Details non disponibles")
df['etat'] = df['etat'].fillna("Etat non spécifié")
df['adresse'] = df['adresse'].fillna("Adresse non disponible")
df['image_link'] = df['image_link'].fillna("lien d'image non disponible")
```

```python
output_file = "cleaned/equipements_cleaned.csv"
```

```python

df.to_csv(output_file, index=False, encoding='utf-!')
```

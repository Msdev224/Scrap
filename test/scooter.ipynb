```python
import pandas as pd
```

```python
df = pd.read_csv("data/brute/expat_scooter.csv")
```

```python
df.head()
```

```python
# Supprimer les colonnes inutiles
df_cleaned = df.drop(columns=["web-scraper-order", "web-scraper-start-url", "scooter-href", "scooter"])
```

```python
df_cleaned.head()
```

```python
df_cleaned.info()
```

```python
df_cleaned["prix"] = df_cleaned["prix"].str.replace(r"[^\d]", "", regex=True).astype(float)
```

```python
df_cleaned = df_cleaned.drop_duplicates()
len(df_cleaned)
```

```python
df_cleaned.head()
```

```python
missing_values = df_cleaned.isna().sum()
missing_values_continuous = missing_values[['annee', 'prix']]
missing_values_categorical = missing_values.drop(['annee', 'prix'])

print("Variables continues avec des valeurs manquantes:")
print(missing_values_continuous[missing_values_continuous > 0])

print("\nVariables catégorielles avec des valeurs manquantes:")
print(missing_values_categorical[missing_values_categorical > 0])
missing_values[missing_values > 0]
```

```python
# Remplir les valeurs manquantes pour les variables continues
df_cleaned['prix'] = df_cleaned['prix'].fillna(df_cleaned['prix'].mean())

# Afficher les valeurs manquantes après remplissage
missing_values_after = df_cleaned.isna().sum()
print(missing_values_after)
```
